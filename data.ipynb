```python
import requests
import time
jsons = []
for y in range(2015, 2021):
    res = requests.get(f"https://serpapi.com/search.json?engine=google_trends&q=акции,мосбиржа,финансы,санкции,банк&data_type=TIMESERIES&csv=true&date={str(y)}-01-01 {str(y+1)}-01-01&api_key=143ce68e42aabdf908dafce79e2edd374fe6e72a2d8a621ae207393aab69ac92&geo=RU")
    time.sleep(2)
    jsons.append(res)
```

```python
for j in jsons:
    print(j.json()["csv"])
```

```python
fout = open("RUS_акции_мосбиржа_финансы_санкции_банк.csv", "w", encoding="utf-8")
head = jsons[0].json()["csv"][2]
fout.write(head+"\n")
for js in jsons:
    print(js.json()["csv"][3:])
    report = js.json()["csv"][3:]
    report = "\n".join(report)
    fout.writelines(report)
    fout.write("\n")
fout.close()
```
